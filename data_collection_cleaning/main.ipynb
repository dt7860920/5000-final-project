# Data Collection from PGA Website
This file is used to scrape data from PGA official website

```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_headers(soup):
    #Collect column names from PGA website
    #Start with empty list
    headers = []

    # "Rounds" header
    rounds = soup.find_all(class_="rounds hidden-small hidden-medium")[0].get_text(strip=True)
    headers.append(rounds)

    # Other stat headers
    stat_headers = soup.find_all(class_="col-stat hidden-small hidden-medium")
    for header in stat_headers:
        headers.append(header.get_text(strip=True))

    return headers


def get_players(soup):
    # Extract player names from a stats page.
    # Start with empty name list
    player_list = []

    # We start collecting from 1 because first one is usually not a player
    players = soup.select("td a")[1:]
    for player in players:
        player_list.append(player.get_text(strip=True))

    return player_list


def get_stats(soup, categories):
    # Collect stats from website
    # Get all numeric data
    stats = soup.find_all(class_="hidden-small hidden-medium")

    stat_list = []
    # Go through the loop to organize data based on how many column of stats we have
    for i in range(0, len(stats) - categories + 1, categories):
        row = []
        for j in range(categories):
            row.append(stats[i + j].get_text(strip=True))
        stat_list.append(row)

    return stat_list

def stats_dict(players, stats):
    # Combine and convert the name and stats list into one, and use index to pull the matching data
    player_dict = {}
    for i, player in enumerate(players):
        player_dict[player] = stats[i]
    return player_dict
```
